```python
import pandas as pd
import numpy as np
```

```python
summer_products = pd.read_csv('C:/Users/Marianna/Desktop/summer_products.csv')
 
```

```python
summer_products = summer_products[['price','rating_count','rating_five_count','origin_country']].groupby(['origin_country']).agg({'price':'mean','rating_count':'sum','rating_five_count':'sum'})
summer_products = pd.DataFrame(summer_products.to_records())
summer_products.sort_values(by=['origin_country'])
```

```python
five_percentage = pd.DataFrame((summer_products['rating_five_count']/summer_products['rating_count']) * 100)

summer_products['five_percentage'] = five_percentage
summer_products = pd.DataFrame(summer_products.to_records())
summer_products
```
